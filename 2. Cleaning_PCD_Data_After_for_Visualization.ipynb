# 0. Import Library

```python
import pandas as pd
```

# 1. Import Data

```python
path = 'cleaned_data/' # Change path here
filename = 'cleaned_pcd_data_after.csv'
```

```python
df = pd.read_csv(path+filename)
df['date_time'] = pd.to_datetime(df['date_time'])
```

```python
df
```

```python
df.info()
```

# 2.1 Create dataframe for visualization

```python
cleaned_df = pd.DataFrame()
```

```python
province_array = df['province'].unique()
month_period = pd.period_range(start='2020-07', end='2022-02', freq='M').astype(str)
```

```python
month_period
```

```python
for province in province_array:
  cleaned_df_new = pd.DataFrame()
  print('Province: '+province)
  selected_df = df[df['province']==province][['date_time', 'lat', 'long', 'PM25']]
  lat_value = selected_df.loc[selected_df.index[0], 'lat']
  long_value = selected_df.loc[selected_df.index[0], 'long']
  cleaned_df_new['date'] = pd.period_range(start='2020-07', end='2022-01', freq='M')
  pm25_mean = []
  for i in range(len(month_period)-1):
    pm25_mean.append(selected_df[(selected_df['date_time']<month_period[i+1]) & (selected_df['date_time']>=month_period[i])]['PM25'].mean())
  cleaned_df_new['PM25'] = pm25_mean
  # Fill first row of each station with 0 if it has NaN value.
  cleaned_df_new.loc[0] = cleaned_df_new.loc[0].fillna(0)
  # Fill missing values with its previous value (avoid using future data)
  cleaned_df_new = cleaned_df_new.interpolate(method = 'pad', limit_direction = 'forward') # Change interpolate method here
  prov_list = [province]*(len(month_period)-1)
  cleaned_df_new['province'] = prov_list
  lat_list = [lat_value]*(len(month_period)-1)
  cleaned_df_new['lat'] = lat_list
  long_list = [long_value]*(len(month_period)-1)
  cleaned_df_new['long'] = long_list
  cleaned_df = pd.concat([cleaned_df, cleaned_df_new], ignore_index=True)
```

```python
cleaned_df
```

```python
cleaned_df = cleaned_df.sort_values(by=['date'])
cleaned_df = cleaned_df.reset_index(drop=True)
```

```python
cleaned_df
```

```python
save_path = 'cleaned_data/'
save_filename = 'cleaned_pcd_data_after_for_visualization.csv'
```

```python
cleaned_df.to_csv(save_path+save_filename, index=False)
```
