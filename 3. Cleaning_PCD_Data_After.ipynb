# 0. Import Library

```python
import pandas as pd
import matplotlib.pyplot as plt
```

# 1. Import Data

```python
path = 'data/PCD Data/Data after 2020-7/' # Change path here
filename = 'PCD data after 2020-7.csv'
```

```python
df = pd.read_csv(path+filename)
```

```python
df.head()
```

# 2. Remove following column: id, stationType, datetime_aq

```python
df = df.drop(['id', 'stationType', 'datetime_aq'], axis=1)
```

```python
df.head()
```

# 3. Change Datetime format

```python
df['date_time'] = df['date_aq'] + ' '+df['time_aq']
df['date_time'] = pd.to_datetime(df['date_time'])
```

```python
df = df.drop(['date_aq', 'time_aq'], axis=1)
```

```python
df.head()
```

```python
df.info()
```

```python
# Change Dataframe column order
cols = df.columns.tolist()
cols = cols[-1:] + cols[:-1]
```

```python
df = df[cols]
```

```python
df.head()
```

# 4. Create province column

```python
arr = df['areaEN'].values
province = []
for i in range(arr.shape[0]):
  area_split = arr[i].split(',')
  prov = area_split[-1].strip()
  # Correct some province
  if (prov == 'Mueang Chiang Mai'):
    prov = 'Chiang Mai'
  if (prov == 'Sakon Nakhon Province'):
    prov = 'Sakon Nakhon'
  if (prov == 'Buriram Province'):
    prov = 'Buriram'
  province.append(prov)
```

```python
df['province'] = province
```

```python
df['province'].nunique()
```

```python
df['province'].unique()
```

```python
df.head()
```

# 5. Impute missing value (Padding with separate stationID)

```python
df.isnull().sum()/len(df) # Percentage of missing value
```

```python
df.info()
```

```python
df['stationID'].nunique() # 151 unique station id
```

```python
unique_station = df['stationID'].unique()
```

```python
interpolate_column = ['PM25', 'PM10', 'NO2', 'SO2', 'CO', 'O3', 'AQI']
cleaned_df = pd.DataFrame()
for station in unique_station:
  print('Station ID: '+station)
  df_selected = df[df['stationID']==station].copy()
  df_selected = df_selected.sort_values(by=['date_time'])
  # Fill first row of each station with 0 if it has NaN value.
  df_selected.loc[df_selected.index[0], interpolate_column] = df_selected.loc[df_selected.index[0], interpolate_column].fillna(0)
  # Fill missing values with its previous value (avoid using future data)
  df_selected[interpolate_column] = df_selected[interpolate_column].interpolate(method = 'pad', limit_direction = 'forward') # Change interpolate method here
  cleaned_df = pd.concat([cleaned_df, df_selected])
```

```python
cleaned_df = cleaned_df.sort_values(by=['date_time'])
cleaned_df = cleaned_df.reset_index(drop=True)
```
